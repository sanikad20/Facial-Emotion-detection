```python
# ==========================================================
# FER2013 Emotion Recognition - Custom CNN (Colab Compatible)
# ==========================================================

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential, model_from_json
from keras.layers import (
    Conv2D, MaxPooling2D, Dense, Flatten,
    Dropout, BatchNormalization, Activation
)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from math import ceil


# ==========================================================
# STEP 1: Load and preprocess dataset
# ==========================================================
def preprocess_data():
    data = pd.read_csv('/content/fer2013.csv')      # <-- CHANGE PATH IF NEEDED
    labels = pd.read_csv('/content/fer2013new.csv') # <-- IMPORTANT: SECOND FILE

    orig_class_names = [
        'neutral', 'happiness', 'surprise', 'sadness',
        'anger', 'disgust', 'fear', 'contempt', 'unknown', 'NF'
    ]

    n_samples = len(data)
    w, h = 48, 48

    y = np.array(labels[orig_class_names], dtype='float32')
    X = np.zeros((n_samples, w, h, 1), dtype='float32')

    for i in range(n_samples):
        X[i] = np.fromstring(data['pixels'][i], dtype=float, sep=' ').reshape((h, w, 1))

    return X, y


# ==========================================================
# STEP 2: Clean and normalize data
# ==========================================================
def clean_data_and_normalize(X, y):
    orig_class_names = [
        'neutral', 'happiness', 'surprise', 'sadness',
        'anger', 'disgust', 'fear', 'contempt', 'unknown', 'NF'
    ]

    y_mask = y.argmax(axis=-1)
    mask = y_mask < orig_class_names.index('unknown')  # remove unknown + NF
    X = X[mask]
    y = y[mask]

    y = y[:, :-2] * 0.1  # reduce unknown/NF columns

    y[:, 0] += y[:, 7]   # merge contempt 竊 neutral
    y = y[:, :7]

    X = X / 255.0

    return X, y


# ==========================================================
# STEP 3: Split dataset
# ==========================================================
def split_data(X, y):
    test_size = ceil(len(X) * 0.1)
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=test_size, random_state=42)
    return x_train, y_train, x_val, y_val, x_test, y_test


# ==========================================================
# STEP 4: Data Augmentation
# ==========================================================
def data_augmentation(x_train):
    datagen = ImageDataGenerator(
        rotation_range=20,
        horizontal_flip=True,
        height_shift_range=0.1,
        width_shift_range=0.1
    )
    datagen.fit(x_train)
    return datagen


# ==========================================================
# STEP 5: Define Model
# ==========================================================
def define_model(input_shape=(48, 48, 1), classes=7):
    num_features = 64

    model = Sequential()

    # Block 1
    model.add(Conv2D(num_features, (3, 3), input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(num_features, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    # Block 2
    model.add(Conv2D(num_features, (3, 3), activation='relu'))
    model.add(Conv2D(num_features, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Block 3
    model.add(Conv2D(2*num_features, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(2*num_features, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))

    # Block 4
    model.add(Conv2D(2*num_features, (3, 3), activation='relu'))
    model.add(Conv2D(2*num_features, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Block 5
    model.add(Conv2D(4*num_features, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(4*num_features, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))

    model.add(Flatten())

    # Dense
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.3))

    model.add(Dense(classes, activation='softmax'))
    return model


# ==========================================================
# STEP 6: Plot
# ==========================================================
def plot_acc_loss(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title("Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title("Loss")
    plt.legend()
    plt.show()


# ==========================================================
# STEP 7: Train + Evaluate
# ==========================================================
def run_model():
    print("Loading data...")
    X, y = preprocess_data()
    X, y = clean_data_and_normalize(X, y)
    x_train, y_train, x_val, y_val, x_test, y_test = split_data(X, y)
    datagen = data_augmentation(x_train)

    print("Train:", x_train.shape, "Val:", x_val.shape, "Test:", x_test.shape)

    model = define_model(input_shape=x_train[0].shape, classes=7)
    model.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    callbacks = [
        tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True),
        tf.keras.callbacks.ModelCheckpoint('/content/best_custom_cnn.h5',
                                           save_best_only=True,
                                           monitor='val_accuracy',
                                           verbose=1)
    ]

    history = model.fit(
        datagen.flow(x_train, y_train, batch_size=64),
        epochs=50,
        validation_data=(x_val, y_val),
        callbacks=callbacks,
        verbose=1
    )

    loss, acc = model.evaluate(x_test, y_test, batch_size=64)
    print(f"\n沐･ Final Test Accuracy: {acc*100:.2f}%")

    plot_acc_loss(history)

    model.save("/content/final_custom_cnn.h5")
    print("Model saved: final_custom_cnn.h5")


# ==========================================================
# RUN EVERYTHING
# ==========================================================
run_model()
```
